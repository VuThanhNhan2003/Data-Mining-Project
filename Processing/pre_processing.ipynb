```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
!pip install liac-arff -q # cài thư viện arff nếu chưa có

import arff
df = pd.read_csv('heart_disease.csv')

# 5 cột categorical cần encode (theo code cũ của bạn)
df.iloc[:,1]  = LabelEncoder().fit_transform(df.iloc[:,1])
df.iloc[:,2]  = LabelEncoder().fit_transform(df.iloc[:,2])
df.iloc[:,6]  = LabelEncoder().fit_transform(df.iloc[:,6])
df.iloc[:,8]  = LabelEncoder().fit_transform(df.iloc[:,8])
df.iloc[:,10] = LabelEncoder().fit_transform(df.iloc[:,10])

# Tạo attributes cho ARFF
attributes = []
for col in df.columns:
    if df[col].dtype == 'object' and col != df.columns[-1]:  
        attributes.append((col, 'NUMERIC'))  
    elif df[col].dtype == 'object':  
        attributes.append((col, sorted(df[col].unique().tolist())))
    else:
        attributes.append((col, 'NUMERIC'))

# Xuất file .arff
with open('heart_disease_preprocessed.arff', 'w', encoding='utf-8') as f:
    arff.dump({
        'relation': 'heart_disease',
        'attributes': attributes,
        'data': df.replace({np.nan: None}).values.tolist()
    }, f)

!dir *.arff #dùng windows
# !ls -la *.arff #dùng linux or mac
```
